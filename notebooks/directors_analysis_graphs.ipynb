```python
def write_graph(graph_file, data, graph_name, desc):
    graph_file.write("<!-- START of graph: " + graph_name + "-->\n\n")
    graph_file.write(data)
    graph_file.write("<br><br>\n\n")
    graph_file.write("<p style='text-align: justify;'>")
    graph_file.write(desc)
    graph_file.write("</p>\n\n")
    graph_file.write("<br><br>\n\n")
    graph_file.write("<hr>\n\n")
    graph_file.write("<br><br>\n\n")
    graph_file.write("<!-- END of graph -->\n\n")

def write_double_table(graph_file, data, data2, graph_name, desc):
    graph_file.write("<!-- START of graph: " + graph_name + "-->\n\n")
    graph_file.write(data)
    graph_file.write(data2)
    graph_file.write("<br><br>\n\n")
    graph_file.write("<p style='text-align: justify;'>")
    graph_file.write(desc)
    graph_file.write("</p>\n\n")
    graph_file.write("<br><br>\n\n")
    graph_file.write("<hr>\n\n")
    graph_file.write("<br><br>\n\n")
    graph_file.write("<!-- END of graph -->\n\n")


target_file = open("graphs_for_directors_v7.html", 'w')
target_file.write('<script src="https://cdn.plot.ly/plotly-latest.min.js"></script>\n\n')
# target_file.write('<link rel="stylesheet" href="https://maxcdn.bootstrapcdn.com/bootstrap/3.3.7/css/bootstrap.min.css" integrity="sha384-BVYiiSIFeK1dGmJRAkycuHAHRg32OmUcww7on3RYdg4Va+PmSTsz/K68vbdEjh4u" crossorigin="anonymous">\n\n')
# target_file.write('<div class="container"> \n \t <div class="row row-centered"> \n\n')
```

```python
def read_from_file(file, actors_amount):

    if actors_amount == 6:

        movies = {}

        with open(file) as csvfile:
            reader = csv.DictReader(csvfile, delimiter="\t")
            for entry in reader:
                movies[
                    entry["title"]
                ] = {
                    "director": entry["director"],
                    "rating": entry["rating"],
                    "votes": entry["votes"],
                    "year": entry["year"],
                    "genre": entry["genre"],
                    "gross": entry["gross"],
                    "budget": entry["budget"],
                    "run-time": entry["run-time"],
                    "actor1": entry["actor1"],
                    "actor1_rank": entry["actor1_rank"],
                    "actor1_sex": entry["actor1_sex"],
                    "actor2": entry["actor2"],
                    "actor2_rank": entry["actor2_rank"],
                    "actor2_sex": entry["actor2_sex"],
                    "actor3": entry["actor3"],
                    "actor3_rank": entry["actor3_rank"],
                    "actor3_sex": entry["actor3_sex"],
                    "actor4": entry["actor4"],
                    "actor4_rank": entry["actor4_rank"],
                    "actor4_sex": entry["actor4_sex"],
                    "actor5": entry["actor5"],
                    "actor5_rank": entry["actor5_rank"],
                    "actor5_sex": entry["actor5_sex"],
                    "actor6": entry["actor6"],
                    "actor6_rank": entry["actor6_rank"],
                    "actor6_sex": entry["actor6_sex"],
                    "plot": entry["plot"]
                }
        return movies

    if actors_amount == 3:

        movies = {}

        with open(file) as csvfile:
            reader = csv.DictReader(csvfile, delimiter="\t")
            for entry in reader:
                movies[
                    entry["title"]
                ] = {
                    "director": entry["director"],
                    "rating": entry["rating"],
                    "votes": entry["votes"],
                    "year": entry["year"],
                    "genre": entry["genre"],
                    "gross": entry["gross"],
                    "budget": entry["budget"],
                    "run-time": entry["run-time"],
                    "actor1": entry["actor1"],
                    "actor1_rank": entry["actor1_rank"],
                    "actor1_sex": entry["actor1_sex"],
                    "actor2": entry["actor2"],
                    "actor2_rank": entry["actor2_rank"],
                    "actor2_sex": entry["actor2_sex"],
                    "actor3": entry["actor3"],
                    "actor3_rank": entry["actor3_rank"],
                    "actor3_sex": entry["actor3_sex"],
                    "plot": entry["plot"]
                }
        return movies
    
movies_main = read_from_file('imdb_dataset.tsv', 6)
```

```python
dir_act = defaultdict(set)

all_Act = set()

for title in movies_main:
    cur_movie_data = movies_main[title]
    cur_dir = cur_movie_data["director"]
    cur_cast = [cur_movie_data["actor1"], cur_movie_data["actor2"], cur_movie_data["actor3"],
                cur_movie_data["actor4"], cur_movie_data["actor5"], cur_movie_data["actor6"]]

    dir_act[cur_dir].update(cur_cast)
    all_Act.update(cur_cast)

for dir1 in dir_act:
    cast1 = dir_act[dir1]
    for dir2 in dir_act:
        if dir1 != dir2:
            cast2 = dir_act[dir2]
            common = cast1.intersection(cast2)
            if len(common) > 0:
                # print "------", dir1,"----", dir2
                for act in common:
                    # add edge between dir1 and dir2
                    pass
            common = []

actors_and_ranks = defaultdict(int)

for title in movies_main:
    actors_and_ranks[movies_main[title]["actor1"]] = movies_main[title]["actor1_rank"]
    actors_and_ranks[movies_main[title]["actor2"]] = movies_main[title]["actor2_rank"]
    actors_and_ranks[movies_main[title]["actor3"]] = movies_main[title]["actor3_rank"]
    actors_and_ranks[movies_main[title]["actor4"]] = movies_main[title]["actor4_rank"]
    actors_and_ranks[movies_main[title]["actor5"]] = movies_main[title]["actor5_rank"]
    actors_and_ranks[movies_main[title]["actor6"]] = movies_main[title]["actor6_rank"]

dir_actors_ranks = defaultdict(lambda: [])

for dir in dir_act:
    act_ranks = []
    for act in dir_act[dir]:
        act_ranks.append(int(actors_and_ranks[act]))
    dir_actors_ranks[dir] = act_ranks

directors_main = {}

for title in movies_main:
    director = movies_main[title]["director"]
    budget = int(movies_main[title]["budget"])
    gross = int(movies_main[title]["gross"])

    if director not in directors_main:
        directors_main[director] = {"name": director,
                                    "gross_all":[gross],
                                    "gross_mean": 0,
                                    "budget_all":[budget],
                                    "budget_mean": 0,
                                    "amount_of_movies":1,
                                    "mean_actors":0}
    else:
        directors_main[director]["gross_all"].append(gross)
        directors_main[director]["budget_all"].append(budget)
        directors_main[director]["amount_of_movies"] += 1


for dir in dir_actors_ranks:
    directors_main[dir]["mean_actors"] = np.mean(dir_actors_ranks[dir])
    directors_main[dir]["gross_mean"] = np.mean(directors_main[dir]["gross_all"])
    directors_main[dir]["budget_mean"] = np.mean(directors_main[dir]["budget_all"])
    # fucking numpy uses 32bit intigers...
    directors_main[dir].update({"gross_total": __builtins__.sum(directors_main[dir]["gross_all"])})
    directors_main[dir].update({"budget_total": __builtins__.sum(directors_main[dir]["budget_all"])})
```

```python
def get_director_stats(dir_data):
    all_dir_gross_total = defaultdict(int)
    all_dir_budget_total = defaultdict(float)
    
    all_dir_budget_mean = defaultdict(float)
    all_dir_gross_mean = defaultdict(float)
    
    all_dir_amount_movies = defaultdict(int)
    
    all_dir_ranks_mean = defaultdict(float)


    for dir in dir_data:
        dir_info = dir_data[dir]
        gross_total = int(dir_info["gross_total"])
        budget_total = float(dir_info["budget_total"])
        gross_mean = float(dir_info["gross_mean"])
        budget_mean = float(dir_info["budget_mean"])
        amount_movies = float(dir_info["amount_of_movies"])
        ranks_mean = float(dir_info["mean_actors"])
        if gross_total < 0:
            print dir_info
        
        all_dir_gross_total[dir] = gross_total
        all_dir_budget_total[dir] = budget_total
        all_dir_budget_mean[dir] = budget_mean
        all_dir_gross_mean[dir] = gross_mean
        all_dir_amount_movies[dir] = amount_movies
        all_dir_ranks_mean[dir] = ranks_mean
        
    return [
            all_dir_gross_total,
            all_dir_budget_total,
            all_dir_budget_mean,
            all_dir_gross_mean,
            all_dir_amount_movies,
            all_dir_ranks_mean ]

result = get_director_stats(directors_main)

res_all_dir_gross_total = result[0]
res_all_dir_budget_total = result[1]
res_all_dir_budget_mean = result[2]
res_all_dir_gross_mean = result[3]
res_all_dir_amount_movies = result[4]
res_all_dir_ranks_mean = result[5]

res_gross_total = [(key, res_all_dir_gross_total[key]) for key in sorted(res_all_dir_gross_total.iterkeys())]
res_gross_mean = [(key, res_all_dir_gross_mean[key]) for key in sorted(res_all_dir_gross_mean.iterkeys())]

res_budge_total = [(key, res_all_dir_budget_total[key]) for key in sorted(res_all_dir_budget_total.iterkeys())]
res_budge_mean = [(key, res_all_dir_budget_mean[key]) for key in sorted(res_all_dir_budget_mean.iterkeys())]

res_amount_movies = [(key, res_all_dir_amount_movies[key]) for key in sorted(res_all_dir_amount_movies.iterkeys())]
res_ranks_mean = [(key, res_all_dir_ranks_mean[key]) for key in sorted(res_all_dir_ranks_mean.iterkeys())]

dir_data = []

for _dir in res_all_dir_gross_total:
    dir_data.append([_dir,                            # 0
                      res_all_dir_gross_total[_dir],  # 1
                      res_all_dir_budget_total[_dir], # 2
                      res_all_dir_budget_mean[_dir],  # 3
                      res_all_dir_gross_mean[_dir],   # 4
                      res_all_dir_amount_movies[_dir],# 5
                      res_all_dir_ranks_mean[_dir] ]) # 6
    
dir_data_sorted = sorted(dir_data, key=itemgetter(0), reverse=False)

hover_text = []

for entry in dir_data_sorted:
    hover_text.append(('<b>Director: {dir}</b><br>'+
                       'Movies count: {count}<br>'+
                       'Actors Rank Mean: {act_m}<br>'+
                       'Budget Mean: {budget_m}<br>'+
                       'Budget Total: {budget_t}<br>'+
                       'Gross Mean: {gross_m}<br>'+
                       'Gross Total: {gross_t}<br>').format(
                                            dir=entry[0],
                                            count=int(entry[5]),
                                            budget_m=hm.intword(entry[3]),
                                            budget_t=hm.intword(entry[2]),
                                            gross_m=hm.intword(entry[4]),
                                            gross_t=hm.intword(entry[1]),
                                            act_m=hm.intword(entry[6])
                        ))
    
amount = len(dir_data)

trace0 = go.Scatter(
    x = [budget for _dir, budget in res_budge_total[:amount]],
    y = [gross for _dir, gross in res_gross_total[:amount]],
    text = hover_text,
    mode='markers',
    marker=go.Marker(size = [amount for _dir, amount in res_amount_movies[:amount]],
                  sizemin = 2,
                  sizeref = 1.2,
                  sizemode ='diameter',
                  colorscale = 'Viridis', #https://github.com/plotly/plotly.js/blob/master/src/components/colorscale/scales.js
                  color = [amount for _dir, amount in res_amount_movies],
                  colorbar = dict(title = 'Amount of <br>movies made'),)
)

layout = go.Layout(
    title = "Directors Statistics All-in-one",
    hovermode="closest",
    
    xaxis=dict(title = "Total Budget USD"),
    yaxis=dict(title = "Total Gross USD"),
    showlegend=False,
    height=700,
)

data = [trace0]
fig = go.Figure(data = data, layout=layout, )
iplot(fig)

write_graph(target_file, 
            plotly.offline.plot(fig, include_plotlyjs=False, output_type='div'),
           "directors all in one",
           "description")
```
